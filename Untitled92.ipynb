###MERGING THE DATASETS FROMING ONE DATASETS

```python
import pandas as pd
ads_campaigns = pd.read_csv('/content/ads_campaigns-sheet1-sourcetable.csv')
customer_table = pd.read_csv('/content/customer_table-sheet1-sourcetable (1).csv')
ga_sessions = pd.read_csv('/content/ga_sessions-sheet1-sourcetable.csv')
orders = pd.read_csv('/content/orders_csv-sheet1-sourcetable.csv')
product = pd.read_csv('/content/poduct_csv-sheet1-sourcetable.csv')
```

```python
# -------------------------------------------------
# Updated Merge Script Based on Your Actual Data
# -------------------------------------------------
import pandas as pd

# Assuming the DataFrames are already loaded as per your first cell
# (If not, uncomment the lines below)
# ads_campaigns = pd.read_csv('/content/ads_campaigns-sheet1-sourcetable.csv')
# customer_table = pd.read_csv('/content/customer_table-sheet1-sourcetable (1).csv')
# ga_sessions = pd.read_csv('/content/ga_sessions-sheet1-sourcetable.csv')
# orders = pd.read_csv('/content/orders_csv-sheet1-sourcetable.csv')
# product = pd.read_csv('/content/poduct_csv-sheet1-sourcetable.csv')

# -------------------------------------------------
# Quick verification (optional – you already saw this)
# -------------------------------------------------
def preview(df, name):
    print(f"\n=== {name} ({df.shape[0]} rows, {df.shape[1]} cols) ===")
    print(df.head(3).to_string(index=False))
    print("Columns:", list(df.columns))

preview(ads_campaigns, "ads_campaigns")
preview(customer_table, "customer_table")
preview(ga_sessions,   "ga_sessions")  # Note: Your output shows order-related cols here!
preview(orders,        "orders")        # Appears identical to ga_sessions in your sample
preview(product,       "product")

# -------------------------------------------------
# Insight from your output:
# - ga_sessions and orders have THE EXACT SAME columns and data (at least in the sample).
#   This might mean ga_sessions is misnamed or contains order-level data with ad_campaign_id.
# - ads_campaigns links via ad_campaign_id (present in ga_sessions/orders as 'ad_campaign_id')
# - No session_id or separate GA metrics (e.g., pageviews) in the provided sample.
# - We'll treat ga_sessions as the base (since it has ad_campaign_id) and merge everything to it.
# -------------------------------------------------

# Start with ga_sessions as the core table (has orders + ad_campaign_id)
df = ga_sessions.copy()

# 1. Merge Product info (on product_id)
df = df.merge(product, on='product_id', how='left', suffixes=('', '_prod'))

# 2. Merge Customer info (on customer_id)
df = df.merge(customer_table, on='customer_id', how='left', suffixes=('', '_cust'))

# 3. Merge Ads Campaign info (on ad_campaign_id)
# Note: Column is 'ad_campaign_id' in ga_sessions, but 'ad_campaign_id' in ads_campaigns
df = df.merge(ads_campaigns, on='ad_campaign_id', how='left', suffixes=('', '_camp'))

# 4. If orders is truly separate (more rows?), merge it too – but in your sample it's identical
#    Uncomment if orders has extra data:
# df = df.merge(orders, on='order_id', how='left', suffixes=('', '_ord'))

# -------------------------------------------------
# Clean up: Drop any fully duplicate columns (e.g., repeated keys after merge)
# -------------------------------------------------
df = df.loc[:, ~df.columns.duplicated()]

# Optional: Clean currency columns (remove $ and commas, convert to float)
currency_cols = ['spend', 'cost_price', 'selling_price', 'price', 'total']
for col in currency_cols:
    if col in df.columns:
        df[col] = df[col].replace('[\$,]', '', regex=True).astype(float)

# Optional: Convert date columns to datetime
date_cols = ['start_date', 'end_date', 'order_date', 'signup_date']
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], format='%m/%d/%Y', errors='coerce')

# -------------------------------------------------
# Final merged dataset
# -------------------------------------------------
print("\n=== FINAL MERGED DATASET ===")
print(f"Shape: {df.shape[0]} rows, {df.shape[1]} columns")
print(df.head(3).to_string(index=False))

# Save to CSV for download/use
df.to_csv('/content/merged_dataset_final.csv', index=False)
print("\nSaved to /content/merged_dataset_final.csv")

# Bonus: Quick summary stats
print("\n=== QUICK INSIGHTS ===")
print(df.describe(include='all'))
print("\nMissing values per column:\n", df.isnull().sum())
```
