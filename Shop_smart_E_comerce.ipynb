##CUSTOMER CHURN PREDICTION

```python
# CELL 2: Feature Engineering for Churn
from datetime import timedelta

# Convert dates
orders['order_date'] = pd.to_datetime(orders['order_date'])
customer_table['signup_date'] = pd.to_datetime(customer_table['signup_date'])

# Define snapshot date (end of data)
snapshot_date = orders['order_date'].max() + timedelta(days=1)

# Aggregate orders per customer
customer_orders = orders.groupby('customer_id').agg(
    total_spent=('total', 'sum'),
    order_count=('order_id', 'count'),
    avg_order_value=('total', 'mean'),
    last_order_date=('order_date', 'max'),
    avg_discount=('discount', 'mean')
).reset_index()

customer_orders['last_order_days'] = (snapshot_date - customer_orders['last_order_date']).dt.days
customer_orders['tenure_days'] = (snapshot_date - customer_table.set_index('customer_id').loc[customer_orders['customer_id'], 'signup_date']).dt.days
customer_orders['order_frequency'] = customer_orders['order_count'] / (customer_orders['tenure_days'] / 30.0)

# Sessions aggregation
# Assuming ga_sessions has columns 'session_id', 'session_duration', 'pages_viewed', 'bounced'
session_agg = ga_sessions.groupby('customer_id').agg(
    session_count=('order_id', 'count'), # Corrected from session_id as it is order_id in ga_sessions
    # Assuming session duration, pages viewed, and bounced are not available in ga_sessions
    # If they exist, replace the following lines with correct column names
    # avg_duration=('session_duration', 'mean'),
    # total_pages=('pages_viewed', 'sum'),
    # bounced_rate=('bounced', 'mean')
).reset_index()

# Merge all
df_churn = customer_orders.merge(customer_table, on='customer_id', how='left')
df_churn = df_churn.merge(session_agg, on='customer_id', how='left').fillna(0)

# Target: churned if no order in last 90 days
df_churn['churned'] = (df_churn['last_order_days'] > 90).astype(int)

# Select features
features = ['tenure_days', 'total_spent', 'order_frequency', 'last_order_days',
            'avg_discount', 'age', 'session_count'] # Removed session_duration, total_pages, bounced_rate
cat_features = ['country', 'gender']

# One-hot encode
df_churn = pd.get_dummies(df_churn, columns=cat_features, drop_first=True)

X = df_churn[features + [col for col in df_churn.columns if col.startswith('country_') or col.startswith('gender_')]]
y = df_churn['churned']

print(f"Churn rate: {y.mean():.2%}")
```
